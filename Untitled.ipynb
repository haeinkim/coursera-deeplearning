```python
from sklearn.datasets import make_blobs

theta, cls = make_blobs(n_samples=400, n_features=2, cluster_std=0.2, centers=[(0.25, 0.25),(0.75, 0.75)])

th = theta * np.pi * 2 
r = np.sin(4 * th)

x1 = r[:,0] * np.cos(th[:,0])
x2 = r[:,1] * np.sin(th[:,1])
print(x1.shape)
plt.subplot(121)
plt.scatter(r[:,0], r[:,1], c=cls)
plt.subplot(122)
plt.scatter(x1, x2, c=cls)
plt.show()
```
